```python
import turtle

turtle.pensize(2)

turtle.pencolor("Black")

turtle.bgcolor("LightBlue")

A = (-600, -400)
B = (-200, -400)
C = (-200, -100)
D = (-600, -100)
E = (-50, -250)
F = (-50, 50)
G = (-230, 250)
H = (-400, 100)

I = (-450, -400)
J = (-350, -400)
K = (-350, -250)
L = (-450, -250)

M = (-120, -250)
N = (-80, -200)
O = (-80, -100)
P = (-120, -150)

# Draw a Dreaming house ^^

# Draw a rectangle
turtle.penup()
turtle.goto(A)
turtle.pendown()
turtle.fillcolor("MediumBlue")
turtle.begin_fill()
turtle.goto(B)
turtle.goto(C)
turtle.goto(D)
turtle.goto(A)
turtle.end_fill()

turtle.penup()
turtle.goto(B)
turtle.pendown()
turtle.fillcolor("Yellow")
turtle.begin_fill()
turtle.goto(E)
turtle.goto(F)
turtle.goto(C)
turtle.goto(B)
turtle.end_fill()

turtle.penup()
turtle.goto(C)
turtle.pendown()
turtle.fillcolor("Orange")
turtle.begin_fill()
turtle.goto(F)
turtle.goto(G)
turtle.goto(H)
turtle.goto(C)
turtle.end_fill()

# Draw a triangle
turtle.penup()
turtle.goto(D)
turtle.pendown()
turtle.fillcolor("Pink")
turtle.begin_fill()
turtle.goto(C)
turtle.goto(H)
turtle.goto(D)
turtle.end_fill()

turtle.penup()
turtle.goto(I)
turtle.pendown()
turtle.fillcolor("LightGreen")
turtle.begin_fill()
turtle.goto(J)
turtle.goto(K)
turtle.goto(L)
turtle.goto(I)
turtle.end_fill()

turtle.penup()
turtle.goto(M)
turtle.pendown()
turtle.fillcolor("Brown")
turtle.begin_fill()
turtle.goto(N)
turtle.goto(O)
turtle.goto(P)
turtle.goto(M)
turtle.end_fill()

# Draw a Tree
A1 = (280, -400)
B1 = (320, -400)
C1 = (320, -250)
D1 = (280, -250)

turtle.penup()
turtle.goto(A1)
turtle.pendown()
turtle.fillcolor("Brown")
turtle.begin_fill()
turtle.goto(B1)
turtle.goto(C1)
turtle.goto(D1)
turtle.goto(A1)
turtle.end_fill()

E1 = (200, -250)
F1 = (400, -250)
G1 = (300, -150)

turtle.penup()
turtle.goto(E1)
turtle.pendown()
turtle.fillcolor("Green")
turtle.begin_fill()
turtle.goto(F1)
turtle.goto(G1)
turtle.goto(E1)
turtle.end_fill()

H1 = (200, -150)
I1 = (400, -150)
J1 = (300, -50)
turtle.penup()
turtle.goto(E1)
turtle.pendown()
turtle.fillcolor("Green")
turtle.begin_fill()
turtle.goto(F1)
turtle.goto(G1)
turtle.goto(E1)
turtle.end_fill()

```
